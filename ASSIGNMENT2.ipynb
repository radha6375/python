 Part E: Mini Project (Mix of All Topics)
  Write a program that: Asks the user for their name and roll number. Takes marks for 5 subjects
 (out of 100). Stores them in a list. Calculates total marks and average. Stores all details in a
 dictionary (name, roll_no, marks, total, average, grade). Prints the final report card in a
 readable format.

```python
# Taking user details
name = input("Enter your name: ")
roll_no = input("Enter your roll number: ")

# Taking marks for 5 subjects
marks = []
print("Enter marks for 5 subjects:")

for i in range(5):
    m = int(input("Enter marks: "))
    marks.append(m)

# Calculating total and average
total = sum(marks)
average = total / 5

# Finding grade
if average >= 90:
    grade = "A"
elif average >= 75:
    grade = "B"
elif average >= 50:
    grade = "C"
else:
    grade = "Fail"

# Storing in dictionary
student = {
    "name": name,
    "roll_no": roll_no,
    "marks": marks,
    "total": total,
    "average": average,
    "grade": grade
}

# Printing report card
print("\n--- REPORT CARD ---")
print("Name:", student["name"])
print("Roll No:", student["roll_no"])
print("Marks:", student["marks"])
print("Total:", student["total"])
print("Average:", student["average"])
print("Grade:", student["grade"])
```
