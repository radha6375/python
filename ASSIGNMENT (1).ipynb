Q 4. Grade Calculator

```python
#Suppose a student has marks in 5 subjects. store them in variables and calculate the average. Based on the average,print the grade
sub1=float(input("Enter marks for subject 1"))
sub2=float(input("Enter marks for subject 2"))
sub3=float(input("Enter marks for subject 3"))
sub4=float(input("Enter marks for subject 4"))
sub5=float(input("Enter marks for subject 5"))
average=(sub1+sub2+sub3+sub4+sub5)/5
if average >=90:
  grade = "A"
elif average >=75:
  grade ="B"
elif average >=50:
  grade ="C"
else:
  grade ="Fail"
print(f"Average Marks:{average:.2f}")
print(f"Grade:{grade}")
```

 Q9. Find the Maximum Number in a List
 Without using the built-in max() function, write code to find the largest number in a list.

```python
numbers=[72,88,14,20,25,45,89]
largest =numbers[0]
for num in numbers:
  if num> largest:
    largest =num
print("The largest numbers in the list is:",largest)
```
