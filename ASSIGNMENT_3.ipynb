```python
#Q2.String Operations Write a program that takes a string from the user and prints: - The string in
 #uppercase - The reversed string - The number of vowels in it
text= input("Enter the string")
print(text.upper())
reverse=text[::-1]
vowels="aeiouAEIOU"
vowel_count = sum(1 for char in text if char in vowels)
print("Reversed:",reverse)
print("Number of vowel",vowel_count)
```

```python
#Q6.While Loop Write a program to print the sum of all even numbers between 1 and 50 using a while
 #loop.
num=1
total=0
while num <=50:
  if num %2==0:
    total +=num
  num +=1
print("Sum of even numbers between 1 and 50 is:",total)
```

```python
#Q13.Functions Write a function find_max(a, b, c) that returns the largest number among three given
    #numbers.
def find_max(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c


print(find_max(10, 25, 15))
```

```python
#Q15.Classes & Objects Create a class Student with attributes name and marks. Add a method
  #display() to print student details. Then create two student objects and display their information.
 #Example Output: Name: Rohan, Marks: 85 Name: Meena, Marks: 92
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print(f"Name: {self.name}, Marks: {self.marks}")


# Creating two student objects
s1 = Student("Rohan", 85)
s2 = Student("Meena", 92)

# Displaying their information
s1.display()
s2.display()
```
